```python
df1["Address"]=df1["Address"]+','+df1["City"]+','+df1["State"]+','+df1["Country"]
df1
```

```python
df1=df1.drop("Addres",1)
df1
```

```python
df1["Coordinates"]=df1["Address"].apply(nom.geocode) #Methode die die iets uitvoert op elke rij
df1
```

```python
df1.Coordinates
```

```python
df1.Coordinates[0]
```

```python
df1.Coordinates[0].latitude
```

```python
df1["Latitude"]=df1["Coordinates"].apply(lambda x: x.latitude) #werkt niet door None value in de kolom
df1
```

```python
df1["Latitude"]=df1["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df1
```

```python
df1["Longitude"]=df1["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df1
```
